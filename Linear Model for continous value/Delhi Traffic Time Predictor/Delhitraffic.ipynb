```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

%matplotlib inline
```

```python
inputs = pd.read_csv("delhi_traffic_features.csv")
target = pd.read_csv("delhi_traffic_target.csv")
print(inputs.head())
print(target.head())
```

```python
inputs = inputs.drop("Trip_ID",axis=1)
target = target.drop("Trip_ID",axis=1)
print(inputs.head())
print(target.head())
```

```python
numerical_cols = inputs.select_dtypes(include=np.number).columns.tolist()
catagorical_cols = inputs.select_dtypes('object').columns.tolist()
numerical_cols
```

```python
numerical_data = inputs[numerical_cols]
catagorical_data = inputs[catagorical_cols]
numerical_data.head()
```

```python
numerical_data.describe()
```

```python
catagorical_data.nunique()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(numerical_data)
```

```python
inputs[numerical_cols] = scaler.transform(inputs[numerical_cols])
inputs
```

```python
categorical_cols = inputs.select_dtypes(include=["object", "category"]).columns.tolist()
print(categorical_cols)
print(type(categorical_cols))
print(categorical_cols)
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

categorical_cols = inputs.select_dtypes(
    include=["object", "category"]
).columns.tolist()

encoder = OneHotEncoder(
    handle_unknown="ignore",
    sparse_output=False
)

encoder.fit(inputs[categorical_cols])

encoder_cols = encoder.get_feature_names_out(categorical_cols)

encoded = encoder.transform(inputs[categorical_cols])

inputs[encoder_cols] = pd.DataFrame(
    encoded,
    columns=encoder_cols,
    index=inputs.index
)

inputs.drop(columns=categorical_cols, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

train_input , test_input,train_target,test_target = train_test_split(inputs,target,test_size=0.3)
print(f"train_input:{train_input.shape}")
print(f"train_target:{train_target.shape}")
print(f"test_input:{test_input.shape}")
print(f"test_target:{test_target.shape}")
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(train_input,train_target)
```

```python
model.coef_
```

```python
pred = model.predict(test_input)
pred
```

```python
from sklearn.metrics import root_mean_squared_error,mean_absolute_error,r2_score
print(f"RMSE:{root_mean_squared_error(test_target,pred)}")
print(f"MAE:{mean_absolute_error(test_target,pred)}")
print(f"R2:{r2_score(test_target,pred)}")
```

```python
from sklearn.linear_model import (
    LinearRegression,Ridge,Lasso,ElasticNet,
    BayesianRidge,ARDRegression,SGDRegressor,
    PassiveAggressiveRegressor,HuberRegressor,
    RANSACRegressor,TheilSenRegressor
)

models = {
    "LinearRegression": LinearRegression(),
    "Ridge": Ridge(),
    "Lasso": Lasso(),
    "ElasticNet": ElasticNet(),
    "BayesianRidge": BayesianRidge(),
    "ARDRegression": ARDRegression(),
    "SGDRegressor": SGDRegressor(max_iter=1000),
    "PassiveAggressive": PassiveAggressiveRegressor(max_iter=1000),
    "Huber": HuberRegressor(),
    "RANSAC": RANSACRegressor(),
    "TheilSen": TheilSenRegressor()
}

results = []

for name,model in models.items():
    try:
        model.fit(train_input,train_target)
        target_pred = model.predict(test_input)
        
        r2 = r2_score(test_target,target_pred)
        rmse = root_mean_squared_error(test_target,target_pred)
        
        results.append((name,r2,rmse))
    except Exception as e:
        print(f"{name} failed :{e}") 

results_df = pd.DataFrame(results,columns=["Model","R2-Score","RMSE"])
results_df = results_df.sort_values(by="RMSE",ascending=False)

print(results_df)           
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=200,
                              random_state=42,
                              max_depth=50,
                              n_jobs=-1)

model.fit(train_input,train_target)
target_pred = model.predict(test_input)

r2 = r2_score(test_target,target_pred)
rmse = root_mean_squared_error(test_target,target_pred)

print(f"Random Forest Regressor r2-Score : {r2} and RMSE : {rmse}")
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=500,
                              random_state=42,
                              max_depth=50,
                              n_jobs=-1)

model.fit(train_input,train_target)
target_pred = model.predict(test_input)

r2 = r2_score(test_target,target_pred)
rmse = root_mean_squared_error(test_target,target_pred)

print(f"Random Forest Regressor r2-Score : {r2} and RMSE : {rmse}")
```
