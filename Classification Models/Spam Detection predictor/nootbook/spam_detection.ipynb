```python
import pandas as pd
```

```python
data = pd.read_csv("../data/spam.csv")
data.head()
```

```python
data.info()
```

```python
data["Category"].value_counts()
```

```python
data["Category"] = data["Category"].map({"ham":0,"spam":1})
data["Category"]
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z0-9 ]', '',text)
    return text

data["Message"] = data["Message"].apply(clean_text)
data["Message"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000)

x = vectorizer.fit_transform(data["Message"])
y = data["Category"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

model = LogisticRegression()

model.fit(x_train,y_train)

pred = model.predict(x_test)

print(f"LogisticRegression:{classification_report(y_test,pred)}")
```

```python
from importlib.util import module_from_spec

from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()

model.fit(x_train,y_train)

pred = model.predict(x_test)

print(f"MultinomialNB:{classification_report(y_test,pred)}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000,binary=True)

x = vectorizer.fit_transform(data["Message"])
y = data["Category"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

from sklearn.naive_bayes import BernoulliNB

model = BernoulliNB()

model.fit(x_train,y_train)

pred = model.predict(x_test)

print(f"BernoulliNB:{classification_report(y_test,pred)}")
```
